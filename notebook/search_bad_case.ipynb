```python
import json
from pathlib import Path

model = "FSRS-4.5"

users = []
ici = []
sizes = []
parameters = []
result_dir = Path(f"../result/{model}")
result_files = result_dir.glob("*.json")
for result_file in result_files:
    with open(result_file, "r") as f:
        result = json.load(f)
        users.append(result["user"])
        ici.append(result[model]["ICI"])
        sizes.append(result["size"])
        parameters.append(result["weights"])
```

```python
import pandas as pd

df = pd.DataFrame({
    "user": users,
    "size": sizes,
    "ici": ici,
    "parameters": parameters
})
```

```python
pd.set_option('display.max_colwidth', None)

df[df["size"]>10000].copy().sort_values(by="ici", ascending=False).head(20)
```

```python
width = 2000

def compute_interval(x, interval_length=width):
    return int((x // interval_length + 0.5) * interval_length)

df['group'] = df['size'].apply(compute_interval)

interval_error = df.groupby('group')['ici'].mean().sort_index().reset_index()

print(interval_error)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(interval_error['group'], interval_error['ici'], width=width)

plt.title('Average Error by Sample Size Interval')
plt.xlabel('Sample Size Interval')
plt.ylabel('Average Error')

plt.show()
```
